# Metode SAW untuk Pemilihan Lahan Sawit 

```python
import pandas as pd
```

Load data

```python
df = pd.read_csv("https://raw.githubusercontent.com/Yulianas24/MetodeSAW/main/file/dataLahan.csv")
df
```

Normalisasi Data

```python
X = df.iloc[:,1:]
Y = df.iloc[:,0]

max = X.max().values

for col in range(len(X.axes[1])):
  for data in range(len(X)):
    X.iloc[data, col]/=max[col]
X
```

Menambahkan Bobot W = [0.35, 0.25, 0.1, 0.2, 0.1]

```python
hasil = []
bobot = [0.35, 0.25, 0.1, 0.2, 0.1]
v1 = (X.iloc[0].values * bobot).sum()

for i in range(len(X)):
  hasil.append((X.iloc[i].values * bobot).sum())
hasil
```

Ranking

```python
data = {'Hasil': hasil, 'Jenis': Y}
hasil_akhir = pd.DataFrame(data = data)
hasil_akhir.sort_values(by = 'Hasil', ascending=False)
```
